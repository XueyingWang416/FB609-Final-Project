# Data Collection

```python
import yfinance as yf
import pandas as pd
import numpy as np
import xlsxwriter

RESULT_DIR = '../Results/'
```

## High Correlation Stocks

```python
high_correlation_stocks = ['EBAY', 'AMZN', 'BBY', 'WMT']

high_correlation_stock_data = yf.download(high_correlation_stocks, 
                      start='2019-01-01', 
                      end='2019-12-31', 
                      progress=False, interval='1d')['Adj Close']
```

```python
high_correlation_stock_data.index = high_correlation_stock_data.index.date
display(high_correlation_stock_data)
```

```python
high_correlation_stock_data.to_excel(RESULT_DIR + 'high_correlation_stocks.xlsx')
high_correlation_stock_data.to_csv(RESULT_DIR + 'high_correlation_stocks.csv')
```

## Low Correlation Stocks

```python
low_correlation_stocks = ['CVX', 'TSLA', 'EA', 'EDU']

low_correlation_stock_data = yf.download(low_correlation_stocks, 
                      start='2019-01-01', 
                      end='2019-12-31', 
                      progress=False, interval='1d')['Adj Close']
```
